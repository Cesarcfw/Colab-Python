```python
import pandas as pd #importa a biblioteca pandas
# criando um Dataframe a partir de uma lista de tuplas:
nomes = ["Vasco", "Claudino", "Clesley"]
idade = [32, 20, 35]
dados = list(zip(nomes, idade))
print(dados)
print()
df = pd.DataFrame(data = dados)
print(df)
```

# Desafio
proposta 4: Dados de musicas de uma plataforma de streaming
* a) Mais ouvidas
* b) Artista com mais musicas
* c) Gênero mais popular
* d) Musica mais recente
* e) Adicionar uma nova coluna com a duração em segundos





```python
# Proposta de exercício de Data Frame: Dados de músicas de uma plataforma de streaming
import pandas as pd

nome_musicas = ["Blinding Lights", "Cruel Summer", "One Dance", "This Is What You Came For", "Sea Of Problems"]
artistas = ["The Weekend", "Taylor Swift", "Drake", "Rihanna", "Glichery"]
genero = ["R&B", "electropop", "HipHop", "House", "Phonk"]
ano_lancamento = [2020, 2019, 2016, 2016, 2023]
duracao_musica = [3.2, 3.0, 4.4, 3.5, 1.5]

dados_musicais = {
    'Música': nome_musicas,
    'Artista': artistas,
    'Gênero': genero,
    'Ano de lançamento': ano_lancamento,
    'Duração (segundos)' : duracao_musica,
}

music_df = pd.DataFrame(dados_musicais)
print(music_df)

# a) Mais ouvidas
print("a) Mais ouvidas: ")
mais_ouvida = music_df.loc[music_df['Ano de lançamento'].idxmax()]['Música']
print(mais_ouvida)
print()
# b) Artista com mais músicas
print("b) Artistas com mais músicas:")
artista_mais_musicas = music_df['Artista'].mode()[0]
print(artista_mais_musicas)
print()
# c) Gênero mais ouvido
print("c) Gênero mais popular:")
genero_mais_popular = music_df['Gênero'].mode()[0]
print(genero_mais_popular)
print()
# d) Música mais recente
print("d) Música mais recente:")
musica_mais_recente = music_df.loc[music_df['Ano de lançamento'].idxmax()]['Música']
print(musica_mais_recente)
print()
# e) Adicionar nova coluna com duração de segundos
print("e) Duração (segundos):")
music_df['Duração (segundos)'] = music_df['Duração (segundos)'] * 60
print(music_df)
print()
display(music_df)
```

Proposta 2: Dados de Estudantes de uma Escola

* a) Média de Notas
* b) Aluno com maior nota em Estrutura de Dados
* c) Aluno mais novo
* d) Aluno com menor nota em Banco de Dados
* e) Adicionar uma nova coluna com a média geral das notas import pandas as pd




```python
aluno = ["Amanda", "Vitu", "Vasco", "César", "Breno"]
notaEd1 = [9, 7, 10, 8, 6]
notaEd2 = [10, 9, 9, 10, 8]
notaBd1 = [8, 5, 3, 6, 7]
notaBd2 = [9, 9, 8, 9, 8]
data_nasc_aluno = [2003, 2006, 1999, 2004, 2001]

dados_alunos = {
    'Alunos' : aluno,
    'Nota ED 1' : notaEd1,
    'Nota ED 2' : notaEd2,
    'Nota BD 1' : notaBd1,
    'Nota BD 2' : notaBd2,
    'Idade' : data_nasc_aluno,
}

estudantes = pd.DataFrame(dados_alunos)
print(estudantes)
print()

#a) Média de Notas
print("a) Média de notas de Estrutura de Dados: ")
media_notas = (estudantes['Nota ED 1'] + estudantes['Nota ED 2']) / 2
print(media_notas)
print()
print("Média de notas de Banco de Dados: ")
media_notas = (estudantes['Nota BD 1'] + estudantes['Nota BD 2']) / 2
print(media_notas)
print()

#b) Aluno com maior nota em Estrutura de Dados
print("b) Aluno com maior nota em Estrutura de Dados: ")
estudantes['Média ED'] = (estudantes['Nota ED 1'] + estudantes['Nota ED 2']) / 2
aluno_maior_nota_ed = estudantes.loc[estudantes['Média ED'].idxmax()]['Alunos']
print(aluno_maior_nota_ed)
print()
print("Aluno com maior nota em Banco de Dados: ")
estudantes['Média BD'] = (estudantes['Nota BD 1'] + estudantes['Nota BD 2']) / 2
aluno_maior_nota_bd = estudantes.loc[estudantes['Média BD'].idxmax()]['Alunos']
print(aluno_maior_nota_bd)
print()

#c) Aluno mais novo
print("c) Aluno mais novo: ")
aluno_mais_novo = estudantes.loc[estudantes['Idade'].idxmax()]['Alunos']
print(aluno_mais_novo)
print()

#d) Aluno com menor nota em Banco de Dados
print("d) Aluno com menor nota em Banco de Dados: ")
estudantes['Média BD'] = (estudantes['Nota BD 1'] + estudantes['Nota BD 2']) / 2
aluno_menor_nota_bd = estudantes.loc[estudantes['Média BD'].idxmin()]['Alunos']
print(aluno_menor_nota_bd)
print()

#e) Adicionar uma nova coluna com a média geral das notas import pandas as pd
print("e) Adicionar uma nova coluna com a média geral das notas: ")
estudantes['Média Geral'] = (estudantes['Nota ED 1'] + estudantes['Nota ED 2'] + estudantes['Nota BD 1'] + estudantes['Nota BD 2']) / 4
print(estudantes)
print()
display(estudantes)
```
