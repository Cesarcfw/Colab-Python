# Avaliação A1

```python
import pandas as pd

#Aluno: César Chaves Fonsêca Watanabe Maluf

#Criação e alimentação de listas para o dicionario
Nome = ["Red Dead Redemption2",
        "God of War: Ragnarok",
        "Grand Theft Auto V",
        "Elden Ring",
        "Horizon Zero Dawn",
        "Anthem"]
#Lista com os nomes dos jogos

Plataforma = ["PS4, XBOXONE, PC",
              "PS5, PS4",
              "PS3, PS4, PS5, XBOX360, XBOXONE, XBOXSERIESX, PC",
              "PC, PS4, PS5, XBOXONE, XBOXSERIESX",
              "PS4, PC",
              "PC, PS4, XBOXONE"]
#Lista com as plataformas em que os jogos foram lançados

Genero = ["Ação/Aventura",
          "Hack and slash",
          "Ação/Aventura, Mundo Aberto",
          "RPG de ação, Soulslike",
          "RPG de ação, Mundo Aberto",
          "Action RPG"]
#Lista com os principais generos dos jogos

Lançamento = ["26/10/2018",
              "09/11/2022",
              "17/09/2013",
              "25/02/2022",
              "28/02/2017",
              "15/02/2019"]
#Lista com as datas de lançamento inicial dos jogos

NotaUser = ["8.8",
            "8.1",
            "8.5",
            "8.0",
            "8.4",
            "4.0"]
#Lista com as notas dadas pela maioria dos usuarios

NotaCrit = ["97",
            "94",
            "97",
            "96",
            "89",
            "59"]
#Lista com as notas dadas pelos criticos

#Criação do dicionario avaliacao_games
avaliacao_games={
    'nomes' :Nome ,
    'plataformas' :Plataforma ,
    'generos' :Genero ,
    'datas' :Lançamento ,
    'nota_geral' :NotaUser ,
    'nota_especialiazada' :NotaCrit
}

#comando para criação do dataframe
games = pd.DataFrame(avaliacao_games)

#I.index para alteração dos numeros de indexação do dataframe
games.index =["1","2", "3", "4", "5", "6"]

```
