```python
# Crie um Python Script que conte quantas vezes cada nome está presente em uma
# lista [’nome1’, ’nome2’, ...] e armazena essa contagem em um dicionário {’nome1’:
# xvezes, ’nome2’: yvezes, ....}.

# Maldito espaço antes do codigo, 2 horas pra fazer funcionar!

def contar_nomes(nomes):

  contar = {}
  for nome in nomes:
    if nome not in contar:
      contar[nome] = 0
    contar[nome] += 1
  return contar

nomes = ["César", "Cleslley", "César", "Victor", "Cleslley", "César"]
contar = contar_nomes(nomes)
print(contar)
```

```python
#Crie um Python Script que realize o mesmo procedimento da questão anterior. No
#entanto, ao invés do conteúdo da lista nomes ser atribuído no próprio script, faça
#uma estrutura de repetição na qual ela leia uma string do usuário e adicione os nomes
#digitados por ele, um de cada vez, na lista nomes. O término da adição de nomes
#deve ser indicado quando o usuário inserir uma string vazia (”).

def conta_nomes():
    lista = []
    while True:
        nome = input("Digite um nome (ou pressione enter para terminar): ")
        if nome == '':
            break
        lista.append(nome)

    dicionario = {}
    for nome in lista:
        if nome in dicionario:
            dicionario[nome] += 1
        else:
            dicionario[nome] = 1
    return dicionario

print(conta_nomes())
```

```python
#Crie um programa que lê uma mensagem do usuário. Com esta mensagem, faça uma
#nova omitindo trocando todos os caracteres de nomes próprios por ’*’. Exemplo:
#se a mensagem for ’Lucas foi ao shopping com Fernanda assistir aquele filme da
#Marvel’, a nova mensagem deverá ser ’***** foi ao shopping com ******** assistir
#aquele filme da ******’. Assuma que um nome próprio sempre começa com letra
#maiúscula e contém apenas letras.

def Censura_maiuscula():
    mensagem = input("Digite uma mensagem: ")
    palavras = mensagem.split()
    nova_mensagem = ""
    for palavra in palavras:
        if palavra[0].isupper():
            nova_mensagem += '*' * len(palavra) + ' '
        else:
            nova_mensagem += palavra + ' '
    return nova_mensagem.strip()

print(Censura_maiuscula())
```
